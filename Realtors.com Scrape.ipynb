```python
def scrolldown(driver, n_scroll):
    body = driver.find_element_by_tag_name('body')
    while n_scroll >= 0:
        body.send_keys(Keys.PAGE_DOWN)
        n_scroll -= 1
    return driver 

d = webdriver.Chrome("./chromedriver")
url = 'https://www.realtor.com/realestateagents/stoneham_ma/sort-recommendations/pg-8'
d.get(url)
time.sleep(5)
realtor_info = []
for x in range(27):
    for y in range(20):
        div = d.find_elements_by_xpath('//div[@class= "agent-name text-bold"]')[y]
        agent_button = div.find_element_by_tag_name('a').text
        div.find_element_by_link_text(agent_button).click();
        time.sleep(6)
        realtor_dictionary = {}
        realtor_dictionary['name'] = agent_button
        try:
            realtor_dictionary['office'] = [el.text for el in d.find_elements_by_xpath('//p[@itemprop= "name"]')]
        except:
            realtor_dictionary['office'] = np.nan
        try:
            realtor_dictionary['address'] =  d.find_element_by_xpath('//p[@class= "agent_address"]').text
        except:
            realtor_dictionary['address'] = np.nan
        try:
            realtor_dictionary['phone_number'] = [el.text for el in d.find_elements_by_xpath('//span[@itemprop= "telephone"]')][0]
        except:
            realtor_dictionary['phone_number'] = np.nan
        realtor_info.append(realtor_dictionary)
        d.back()
        time.sleep(5)
        print(y)
    next_button = d.find_element_by_xpath('//a[@class= "next"]')
    next_button.click()
    time.sleep(2)
```

```python
realtor_df = pd.DataFrame(realtor_info)
```

```python
test_df = realtor_df
test_df.fillna('N/A',inplace=True)
test_df.at[278, 'address'] = 'Suite 109\nActon, MA 01718'
test_df.at[287, 'address'] = '424 Salem St\nMedford, MA 02155'
```

```python
def convert_list_to_string(org_list, seperator=' '):
    return seperator.join(org_list)

for index, row in test_df.iterrows():
    if (len(row[1])) < 5:
        row[1] = convert_list_to_string(row[1])
    elif (len(row[1])) >= 5:
        row[1] = row[1][3]

address_list = []
state_list = []
city_list = []
zip_list = []
for index, row in test_df.iterrows():
    if row[2] != 'N/A':
        address_list.append(row[2].split("\n")[0])
        city_state_zip = row[2].split("\n")[1] 
        city_state = ''.join([i for i in city_state_zip if not i.isdigit()])
        zip_list.append(re.sub('[^0-9]','', city_state_zip))
        state_list.append(city_state.split(',')[0].strip())
        city_list.append(city_state.split(',')[1].strip())
    else:
        address_list.append('N/A')
        state_list.append('N/A')
        city_list.append('N/A')
        zip_list.append('N/A')
test_df['address'] = address_list
test_df['city'] = city_list
test_df['state'] = state_list
test_df['zip_code'] = zip_list
```
